```python
#Modification 5: Lors de notre exploration nous avons remarqué que  les prix (currency_buyer) étaient dans des devises différentes, ce qui rend compliquée une comparaison
#Nous allons les convertir en EUR en utilisant une bibliothèque de conversion des devises.
!pip install forex_python
```

```python
from forex_python.converter import CurrencyRates
c = CurrencyRates()

# On Convertit les montants en euros (EUR) en utilisant la colonne 'currency_buyer'
def convert_to_eur(row):
    try:
        rate = c.get_rate(row['currency_buyer'], 'EUR')
        price_eur = row['price'] * rate
        retail_price_eur = row['retail_price'] * rate
        return pd.Series({'price_eur': price_eur, 'retail_price_eur': retail_price_eur})
    except:
        return pd.Series({'price_eur': None, 'retail_price_eur': None})

# On applique la conversion des prix
df_eur = df.apply(convert_to_eur, axis=1)

# On ajoute les colonnes converties au  DataFrame d'origine df:
df[['price_eur', 'retail_price_eur']] = df_eur[['price_eur', 'retail_price_eur']]
```

```python
#Je supprime les colonnes currency_buyer, price et retail_price (remplacées par les deux autres nouvelles colonnes price_eur et retail_price_eur)
df=df.drop(['currency_buyer', 'price', 'retail_price'], axis=1)
```

```python
#Vérification des doublons par combinaison des colonnes : product_id, price_eur et retail_price_eur
duplicates = df.duplicated(subset=['price_eur', 'retail_price_eur', 'product_id']).sum()
duplicates
```

```python
#suppression des doublons
df.drop_duplicates(subset=['price_eur', 'retail_price_eur', 'product_id'], inplace=True)
```

```python
#Modification 6 : concerne la gestion des nans
#Nous avons vu que certaines colonnes possèdent des valeurs manquantes, nous créons une fonction pour ne voir que les colonnes avec valeurs manquantes
def valeur_manquante(df) :
  count=0
  for i in df.columns:
    if df[i].isna().sum() >0:
      count=1
      print(f' "{i}" : {df[i].isna().sum()} valeurs manquantes')
valeur_manquante(df)
```

```python
#Explorons les lignes de ces valeurs manquantes
df.loc[pd.isnull(df['rating_five_count']), ['rating','rating_count','rating_four_count','rating_three_count','rating_two_count', 'rating_one_count','origin_country']]
```

```python
#Les valeurs manquantes pour les rating(one à five)_count sont liées à l'absence de note , il pourrait être logique de remplacer les valeurs manquantes par 0.
df[['rating_five_count','rating_four_count','rating_three_count','rating_two_count', 'rating_one_count']]=df[['rating_five_count','rating_four_count','rating_three_count','rating_two_count', 'rating_one_count']].replace(np.nan, 0)
```

```python
#Pour orgin_country, nous pouvons supposer que soit les vendeurs n'ont pas précisé le pays d'origine des produits ou soit le pays n'est pas dans la liste.
#Nous remplacerons les valeurs manquantes d'origin_country par 'Other'.
df['origin_country']=df['origin_country'].replace(np.nan, 'Other')
```
